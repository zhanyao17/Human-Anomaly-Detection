```python
import cv2
import os

def extract_frames(video_path, frames_dir):
    # Open the video file
    video_capture = cv2.VideoCapture(video_path)

    # Create the frames directory if it doesn't exist
    os.makedirs(frames_dir, exist_ok=True)

    # Initialize frame count
    frame_count = 0

    # Read the video frame by frame
    while True:
        # Capture frame-by-frame
        ret, frame = video_capture.read()

        # Check if the video has ended
        if not ret:
            break

        # Save the frame as an image file
        frame_filename = os.path.join(frames_dir, f"frame_{frame_count:04d}.jpg")
        cv2.imwrite(frame_filename, frame)

        # Increment frame count
        frame_count += 1

    # Release the video capture object
    video_capture.release()
```

```python
import cv2
import numpy as np
import os

def video_enhance(video_path, des_path, output_filename):
    # Open the video file
    cap = cv2.VideoCapture(video_path)

    # Check if the video opened successfully
    if not cap.isOpened():
        print("Error: Could not open video.")
        return

    # Get video properties
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)

    # Define the codec and create a VideoWriter object to save the processed video
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(os.path.join(des_path, output_filename), fourcc, fps, (frame_width, frame_height))

    # Create CLAHE object
    clahe = cv2.createCLAHE(clipLimit=0.5, tileGridSize=(5, 5))

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Convert frame to LAB color space
        lab = cv2.cvtColor(frame, cv2.COLOR_BGR2LAB)
        l_channel, a, b = cv2.split(lab)

        # Apply CLAHE to L-channel
        cl = clahe.apply(l_channel)

        # Merge CLAHE enhanced L-channel with the a and b channel
        limg = cv2.merge((cl, a, b))

        # Convert image from LAB to BGR color space
        enhanced_frame = cv2.cvtColor(limg, cv2.COLOR_LAB2BGR)

        # Write the enhanced frame to the output video
        out.write(enhanced_frame)

    # Release everything when job is finished
    cap.release()
    out.release()
    cv2.destroyAllWindows()

# perform enhace for all video set
base_path = "/content/dataset_action_split/train"
base_path_list = [d for d in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, d))]

for dir in base_path_list:
  subdir_path = os.path.join(base_path,dir)
  for vid in os.listdir(subdir_path):
      video_path = os.path.join(subdir_path,vid)
      des_path = '/content/enhanced_contrast'+dir
      video_enhance(video_path,des_path,vid)
```

```python
import cv2
import os

def resizeVideo(video_path,output_path,video_name):
    '''
    video_path: original video path
    outpath: /content/resized_train/Fall Down/
    video_name: video_0.avi
    '''

    # Open the video file
    cap = cv2.VideoCapture(video_path)

    # Check if the video opened successfully
    if not cap.isOpened():
        print("Error: Could not open video.")
        exit()

    # Define the desired width and height for resizing
    new_width = 224  # New width
    new_height = 224  # New height

    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    output_final = os.path.join(output_path,video_name)
    out = cv2.VideoWriter(output_final, fourcc,
                          30.0, (new_width, new_height))

    # Read and resize each frame, then
    # write it to the output video-
    while cap.isOpened():
        ret, frame = cap.read()

        if not ret:
            break

        # Resize the frame
        resized_frame = cv2.resize(frame,
         (new_width, new_height))

        # Write the resized frame to the output video
        out.write(resized_frame)

    # Release the video capture and writer objects
    cap.release()
    out.release()

# perform resize for all video set
base_path = "/content/dataset_action_split/train"
base_path_list = [d for d in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, d))]

for dir in base_path_list:
  subdir_path = os.path.join(base_path,dir)
  for vid in os.listdir(subdir_path):
      video_path = os.path.join(subdir_path,vid)
      des_path = '/content/resized'+dir
      video_enhance(video_path,des_path,vid)
```

## Drop video that have frame < 30 or > 70

```python
def check_num_frame(vidPath):
  video = cv2.VideoCapture(vidPath)
  # Get total number of frames
  frame_count = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
  return frame_count

def drop_short_vid(filePath):
  for fn in os.listdir(filePath):
    video_path = os.path.join(filePath, fn)
    frame_count = check_num_frame(video_path)
    if frame_count < 30 or frame_count > 70:
      print(f'removed: {video_path}')
      os.remove(video_path)


# perform resize for all video set
base_path = "/content/dataset_action_split/train"
base_path_list = [d for d in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, d))]

for dir in base_path_list:
  subdir_path = os.path.join(base_path,dir)
  drop_short_vid(subdir_path)
```

## Duplicate frames

```python
import cv2
import numpy as np

def increase_frame_count_avi(input_file, output_file, new_frame_count):
    # Open the input AVI file
    cap = cv2.VideoCapture(input_file)

    # Get video properties
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    num_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    # Read all frames and append to the frame list
    frames = []
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frames.append(frame)

    cap.release()

    # Duplicate the last frame to reach the new_frame_count
    if new_frame_count > num_frames:
        last_frame = frames[-1]
        num_duplications = new_frame_count - num_frames
        for _ in range(num_duplications):
            frames.append(last_frame)

    # Write the new video with increased frame count to output_file
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(output_file, fourcc, fps, (frame_width, frame_height))

    for frame in frames:
        out.write(frame)

    out.release()

fp = '/content/dataset_action_split/train' # file path
dp = '/content/dataset_resized/duplicate_train' # destination path

for fn in os.listdir(fp):
  subdir_path = os.path.join(fp,fn)
  for vid in os.listdir(subdir_path):
    vp = os.path.join(subdir_path, vid) # video path
    des_vp = os.path.join(dp,fn) # destination video path
    increase_frame_count_avi(vp,des_vp,40)
```
